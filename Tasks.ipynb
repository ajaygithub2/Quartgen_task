### Reading Data

```python
df = pd.read_excel("Sample StockMarket Data.xlsx")
```

### Task 1: Classify data based on expiry dates

```python
# As the data is only available for BANKNIFTY, I went ahead and selected first 17 characters which covers text upto date, the code is static
# but can be made dynamic according to requirements

unique_dates = df["Ticker"].apply(lambda x: x[0:16]).unique()


# Creating empty dictionary to get dataframes for different dates in a dictionary
dict_of_df = {}


# New dataframe created and assigned to a key in dictionary
for date in unique_dates:
    key_name = date
    dict_of_df[key_name] = df.loc[df["Ticker"].apply(lambda x: x.__contains__(f"{date}"))]
```

### Task 2: Create dataframes for different time intervals

```python
# 3 minute interval dataframe
three_minute_interval = df.loc[df["Time"].apply(lambda x: int(x.minute) % 3 == 0)]

# 5 minute interval dataframe
five_minute_interval = df.loc[df["Time"].apply(lambda x: int(x.minute) % 5 == 0)]

# 15 minute interval dataframe
fifteen_minute_interval = df.loc[df["Time"].apply(lambda x: int(x.minute) % 15 == 0)]

# 60 minute interval dataframe
sixty_minute_interval = df.loc[df["Time"].apply(lambda x: int(x.minute) == 0)]
```
